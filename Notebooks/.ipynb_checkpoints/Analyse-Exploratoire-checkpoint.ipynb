```python
#  Enregistrer le nombre de lignes avant nettoyage
rows_before = len(building_consumption)
print("Nombre de lignes avant nettoyage :", rows_before)

#  Détection des incohérences
current_year = datetime.now().year

mask_surface = building_consumption["PropertyGFATotal"] < 0
mask_year = building_consumption["YearBuilt"] > current_year
mask_floors = (building_consumption["NumberofFloors"] <= 0) | (building_consumption["NumberofFloors"] > 200)
mask_energy = (building_consumption["Electricity(kWh)"] < 0) | (building_consumption["NaturalGas(therms)"] < 0)
mask_emissions = building_consumption["TotalGHGEmissions"] < 0

#  Mask global
invalid_mask = mask_surface | mask_year | mask_floors | mask_energy | mask_emissions

# Compter les anomalies
invalid_count = invalid_mask.sum()
print("Nombre de lignes incohérentes détectées :", invalid_count)

# Créer la version nettoyée
building_clean = building_consumption[~invalid_mask].copy()

# Enregistrer le nombre après nettoyage
rows_after = len(building_clean)
print("Nombre de lignes après nettoyage :", rows_after)

#  Pourcentage supprimé
print("Pourcentage supprimé :", round((rows_before - rows_after) * 100 / rows_before, 2), "%")
```

---
##  Nettoyage des données et détection d’anomalies

Cette étape vise à identifier et supprimer les lignes incohérentes du dataset `building_consumption` afin d’assurer la qualité des analyses futures.

### Critères d’incohérence détectés
- **Surface totale (`PropertyGFATotal`)** inférieure à 0
- **Année de construction (`YearBuilt`)** supérieure à l’année actuelle
- **Nombre d’étages (`NumberofFloors`)** ≤ 0 ou > 200
- **Consommation énergétique (`Electricity(kWh)` ou `NaturalGas(therms)`)** négative
- **Émissions de GES (`TotalGHGEmissions`)** négatives

###  Résultats du nettoyage
- **Nombre de lignes avant nettoyage** : 3376  
- **Nombre de lignes incohérentes détectées** : 17  
- **Nombre de lignes après nettoyage** : 3359  
- **Pourcentage supprimé** : 0.5 %

###  Étapes réalisées
- Création de masques booléens pour chaque critère
- Agrégation des masques pour identifier les lignes invalides
- Suppression des lignes incohérentes
- Création d’un nouveau DataFrame `building_clean` propre

---

Ce nettoyage garantit que les données utilisées pour la modélisation ou la visualisation sont logiques et exploitables.

----

# Etape 1 : Compréhension des colonnes
## Colonnes d’identification et localisation
- **OSEBuildingID**: identifiant unique du bâtiment (clé primaire pour relier les données).
- **PropertyName / Address / City / State / ZipCode** : informations d’adresse, utiles pour géolocaliser ou croiser avec d’autres bases.
- **TaxParcelIdentificationNumber** : numéro cadastral, identifiant administratif.
- **CouncilDistrictCode / Neighborhood** : localisation politique et géographique (district, quartier).
- **Latitude / Longitude** : coordonnées géographiques précises pour cartographie et analyses spatiales.
## Colonnes temporelles
- **DataYear** : année de collecte des données.
- **YearBuilt** : année de construction du bâtiment, utile pour analyser l’impact de l’âge sur la consommation.
- **YearsENERGYSTARCertified** : nombre d’années où le bâtiment a été certifié ENERGY STAR.
## Caractéristiques physiques du bâtiment
- **BuildingType / PrimaryPropertyType** : type de bâtiment (résidentiel, bureau, école, etc.).
- **NumberofBuildings / NumberofFloors** : nombre de bâtiments et d’étages sur la parcelle.
- **PropertyGFATotal** : surface totale (Gross Floor Area).
- **PropertyGFAParking / PropertyGFABuilding(s)** : surfaces dédiées au parking et aux bâtiments.
- **LargestPropertyUseType / LargestPropertyUseTypeGFA** : usage principal et surface correspondante.
- **SecondLargestPropertyUseType / ThirdLargestPropertyUseType (+ GFA)** : usages secondaires et surfaces associées.
## Colonnes de performance énergétique
- **ENERGYSTARScore** : score de performance énergétique (0–100).
- **SiteEUI(kBtu/sf)** : Energy Use Intensity (EUI) sur site, consommation par surface.
- **SiteEUIWN(kBtu/sf)** : EUI normalisé selon conditions météo.
- **SourceEUI(kBtu/sf)** : EUI basé sur l’énergie primaire (avant conversion).
- **SourceEUIWN(kBtu/sf)** : EUI primaire normalisé météo.
- **SiteEnergyUse(kBtu) / SiteEnergyUseWN(kBtu)** : consommation totale d’énergie sur site (normalisée ou non).
- **SteamUse(kBtu)** : consommation de vapeur.
- **Electricity(kWh / kBtu)** : consommation électrique en kWh et en kBtu.
- **NaturalGas(therms / kBtu)** : consommation de gaz naturel en therms et en kBtu.
## Colonnes environnementales
- **TotalGHGEmissions** : émissions totales de gaz à effet de serre (CO₂ équivalent).
- **GHGEmissionsIntensity** : intensité des émissions rapportée à la surface (kg CO₂e/m²).
## Colonnes de contrôle et qualité
- **DefaultData** : indique si des valeurs par défaut ont été utilisées.
- **Comments** : remarques éventuelles.
- **ComplianceStatus** : statut de conformité réglementaire.
- **Outlier** : flag indiquant si le bâtiment est considéré comme atypique.

----

```python
# 1. Sélection des colonnes à  supprimer
# Colonnes inutiles pour l’analyse (administratives)
administrative_cols = [
    "TaxParcelIdentificationNumber",
    "PropertyName",
    "Address",
    "City",
    "State",
    "Comments"
]

#  Colonnes énergétiques à risque (target leakage)
energy_cols = [
    "SiteEUI(kBtu)", 
    "SourceEUI(kBtu)",
    "WeatherNormalizedSiteEUI(kBtu)",
    "WeatherNormalizedSourceEUI(kBtu)",
    "Electricity(kWh)",
    "Electricity(kBtu)",
    "NaturalGas(therms)",
    "NaturalGas(kBtu)",
]

# 2. Colonnes trop vides ou inutilisables
cols_inutilisables = [
    "YearsENERGYSTARCertified"
]

# Colonnes finales à supprimer
cols_to_drop = administrative_cols + energy_cols + cols_inutilisables

building_clean = building_consumption.drop(columns=cols_to_drop, errors="ignore")

print("Colonnes supprimées :")
for col in cols_to_drop:
    if col in building_consumption.columns:
        print("  -", col)
```

###  Analyse des relations entre la cible et les variables numériques

Nous avons généré des **scatterplots** pour explorer les relations entre la variable cible `SiteEUIWN(kBtu/sf)` (intensité énergétique normalisée par la météo) et plusieurs variables numériques du dataset.

####  Variables structurelles

####  Variables énergétiques (EDA uniquement)
- `Electricity(kWh)` et `NaturalGas(therms)` : forte corrélation attendue avec la cible, mais **non utilisées pour le modèle** pour éviter le **target leakage**.  
- `TotalGHGEmissions` et `GHGEmissionsIntensity` : corrélations visibles, mais également exclues du modèle.

 **Échantillon utilisé** : 300 lignes aléatoires pour éviter la surcharge visuelle.  
 Ces visualisations permettent d’identifier les variables explicatives pertinentes et de repérer les effets non linéaires ou les outliers.

Souhaites-tu que je t’aide à calculer les **coefficients de corrélation linéaire** pour ces variables ou à tester des transformations (log, carré, etc.) pour améliorer la lisibilité ?

```python
# --- Target selon recommandations ---
target_col = "SiteEUIWN(kBtu/sf)"

# --- Supprimer les doublons de colonnes dans le dataset ---
building_clean = building_clean.loc[:, ~building_clean.columns.duplicated()]

# --- Liste des colonnes énergétiques à exclure ---
energy_cols = [
    "SiteEnergyUse(kBtu)", "Electricity(kWh)", "NaturalGas(therms)",
    "SourceEUI(kBtu/sf)", "SourceEUIWN(kBtu/sf)",
    "TotalGHGEmissions", "GHGEmissionsIntensity"
]

# --- Colonnes numériques sans leakage ---
numeric_cols = building_clean.select_dtypes(include=['float64', 'int64']).columns
numeric_cols = [c for c in numeric_cols if c not in energy_cols]

# --- Vérifier que la target est bien là ---
if target_col not in building_clean.columns:
    raise ValueError(f"{target_col} n'existe pas dans le dataset")

# --- Construire un dataset propre ---
df_corr = building_clean[numeric_cols + [target_col]].copy()
df_corr = df_corr.select_dtypes(include=["float64", "int64"])

# --- Matrice de corrélation ---
corr_matrix = df_corr.corr()

# --- Résoudre le problème : récupérer UNIQUEMENT la 1ère colonne avec ce nom ---
target_corr_series = corr_matrix.loc[:, target_col]

# --- S'assurer que c'est une Series ---
if isinstance(target_corr_series, pd.DataFrame):
    target_corr_series = target_corr_series.iloc[:, 0]

# --- Trier ---
corr_target = target_corr_series.sort_values(ascending=False)

# --- Top variables ---
top_vars = corr_target.drop(target_col).head(10).index

# --- Sous-matrice ---
subset_corr = corr_matrix.loc[top_vars.tolist() + [target_col], top_vars.tolist() + [target_col]]

# --- Heatmap ---
plt.figure(figsize=(10, 8))
sns.heatmap(subset_corr, annot=True, cmap="coolwarm", square=True)
plt.title("Top 10 des variables les plus corrélées avec la cible")
plt.tight_layout()
plt.show()

```

###  Corrélation avec la cible `SiteEUIWN(kBtu/sf)`

Nous avons calculé les coefficients de corrélation entre la cible `SiteEUIWN(kBtu/sf)` (intensité énergétique normalisée par la météo) et les variables numériques du dataset, en excluant toutes les variables énergétiques susceptibles de provoquer du **target leakage** :

####  Variables exclues pour éviter le leakage :
- `SiteEnergyUse(kBtu)`
- `Electricity(kWh)`
- `NaturalGas(therms)`
- `SourceEUI(kBtu/sf)`
- `SourceEUIWN(kBtu/sf)`

####  Méthodologie :

- Visualisation via une **heatmap annotée**
